```python
#Libreria que contiene sets de datos financieros
import yfinance as yf
df_dolar = yf.download("KIMBERA.MX").reset_index()[["Date","Close"]]
#para ver mas información ,ver la pag de yahoo finance para bajar los datos
```

```python
df_dolar
```

```python
plt.figure(figsize=(20,4))
x = df_dolar['Date']
y = df_dolar['Close']
plt.plot(x,y)
#tenemos fecha y precio(valores) de cierrre del dollar
```

```python
df.rename(columns={'date':"Date"},inplace=True)
```

diferente tipo de variable para hacer un merge con df y df_dolar

```python
df.Date.dtypes
```

```python
df_dolar.Date.dtypes
#df_dolar["Date"].info()
```

```python
#transformar la columna Date a formato fecha para poder unir con el otro Df
pd.to_datetime(df["Date"],format='%Y-%m-%d')
```

```python
#igualamos para que tenga el mismo formato(que se guarden los cambos)
df["Date"] = pd.to_datetime(df["Date"],format='%Y-%m-%d')
```

```python
df.Date.dtypes#verificamos el tipo de dato
```

```python
#hacer un merge entre tablas para relacionar el precio del aguacate
df = df.merge(df_dolar,on=["Date"])
```

```python
df_pesos = df
```

```python
df_pesos.head()
```

```python
df_pesos.rename(columns={'Close':"peso_dollar"},inplace=True)
```

```python
#si queremos saber el precio del agucate en pesos
#necesitamos multiplicar las columnas
df_pesos['pesos'] = df_pesos['average_price'] * df_pesos['peso_dollar']
```

```python
df_pesos.tail()
```

```python
plt.figure(figsize=(20,5))
x = df_pesos[(df_pesos.geography == "Las Vegas") & (df_pesos.type == "organic")].sort_values('Date')["Date"] 
y = df_pesos[(df_pesos.geography == "Las Vegas") & (df_pesos.type == "organic")].sort_values('Date')["pesos"]
plt.plot(x,y)
#tenemos su evolucion en pesos del aguacate desde el 2019 a 2020 en las Vegas
#aumenta considerablemente a traves de los años
```
